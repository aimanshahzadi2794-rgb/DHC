```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

sns.set_theme()
```

```python
iris = load_iris()

df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

df["species"] = iris.target
df["species"] = df["species"].map({
    0: "setosa",
    1: "versicolor",
    2: "virginica"
})

df.head()
```

```python
print("Dataset Shape:", df.shape)
print("Column Names:", df.columns.tolist())
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=df,
    x="sepal length (cm)",
    y="petal length (cm)",
    hue="species"
)
plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()
```

```python
df.hist(figsize=(10, 8))
plt.suptitle("Feature Distributions in Iris Dataset")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.drop(columns=["species"]))
plt.title("Box Plot of Iris Features")
plt.show()
```

```python
sns.pairplot(df, hue="species")
plt.show()
```

```python
# Explanation of Results and Final Insights

# After exploring the Iris dataset, we observed that the dataset contains 150 records and 5 columns. There are four 
# numerical features: sepal length, sepal width, petal length, and petal width, along with the species column.

# The scatter plot showed that petal length is useful for separating different Iris species. Setosa can be clearly 
# separated from Versicolor and Virginica. The histograms showed the distribution of each feature, and the box plots 
# helped identify the spread of values and possible outliers.

# Overall, petal length and petal width appear to be the most useful features for distinguishing between Iris flower 
# species.
```
